# Data Augmentation Demonstration

This notebook simply displays the data augmentation techniques and practices used for the thesis research.

The notebook makes use of helper functions implemented by the researchers in the `OCR_Modles_Augmentation.py` file.

# Importing

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from OCR_Models_Augmentation import (
    rotate_image,
    translate_image,
    scale_image,
    flip_image,
    affine_transform_image,
    perspective_transform,
    adjust_brightness,
    adjust_constrant,
    add_gaussian_noise,
    add_gaussian_blur,
    add_color_jitter,
    sharpen_image,
    erode_image,
    dilate_image,
    open_image,
    close_image,
    tophat_image
)
```

# Helper Function
Function below will allow us to print the original and augmented data side by side for easier analysis.

```python
def show_side_by_side(original, augmented, title):
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(augmented, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")

    plt.show()
```

# Load Image
Feel free to replace with any other image within `data_png` directory.

```python
img = cv2.imread("data_png/data_1.png")
```

# Augmenting Data

## Photometric Transformations

```python
show_side_by_side(img, adjust_brightness(img, 40), "Brightness +40")
show_side_by_side(img, adjust_constrant(img, 1.4), "Contrast 1.4")
show_side_by_side(img, add_gaussian_noise(img), "Gaussian Noise")
show_side_by_side(img, add_gaussian_blur(img), "Gaussian Blur")
show_side_by_side(img, add_color_jitter(img), "Color Jitter")
show_side_by_side(img, sharpen_image(img), "Sharpen")
```

# Geometric Transformations

```python
show_side_by_side(img, rotate_image(img, 20), "Rotation 20")
show_side_by_side(img, translate_image(img, 30, 50), "Translation (30,50)")
show_side_by_side(img, scale_image(img, 1.2), "Scale 1.2x")
#show_side_by_side(img, flip_image(img, "horizontal"), "Horizontal Flip")

pts2 = np.float32([[70, 100], [220, 80], [100, 250]])
show_side_by_side(img, affine_transform_image(img, pts2), "Affine Transform")

show_side_by_side(img, perspective_transform(img), "Perspective Transform")
```

# Morphological Transformations

```python
# Erosion
show_side_by_side(img, erode_image(img), "Erosion")
show_side_by_side(img, dilate_image(img), "Dilation")
show_side_by_side(img, open_image(img), "Opening")
show_side_by_side(img, close_image(img), "Closing")
show_side_by_side(img, tophat_image(img), "Tophat")
```
